```python
import polars as pl

suffixes = pl.read_csv('data/morphynet/suffixes.csv')['suffix']
prefixes = pl.read_csv('data/morphynet/prefixes.csv')['prefix']
```

```python
from tokenizer import train_mopiece

train_mopiece('tokenizer', ['data/zen.txt'], prefixes, suffixes, 91, min_stem=3)
```

```python
from tokenizer import MOPiece

mopiece = MOPiece('tokenizer')

with open('data/zen.txt', mode='r', encoding='utf8') as f:
    tokens = mopiece.encode(f.readlines())

tokens
```

```python
print('\n'.join(mopiece.decode(tokens)))
```

```python
with open('data/zen.txt', mode='r', encoding='utf8') as f:
    prefix_tokens, spm_tokens, suffix_tokens = mopiece.encode(f.readlines(), morpheme_first=True)

prefix_tokens, spm_tokens, suffix_tokens
```

```python
print('\n'.join(mopiece.decode((prefix_tokens, spm_tokens, suffix_tokens), morpheme_first=True)))
```
